```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),

]
```

```python
df = pd.DataFrame(prices)
```

```python
df = pd.DataFrame(prices, columns=["month", "price"])
df = df.set_index("month")
```

```python
priceUSD = df['price'].apply(lambda price: price / 4)
df['priceUSD'] = priceUSD
plt.plot(df['priceUSD'],'r--')
plt.ylim(0, 2)
plt.xlabel('Months')
plt.grid()
plt.title('Price of goods (USD)')
plt.show()
```
